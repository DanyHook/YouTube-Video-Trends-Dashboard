```python
# Cargar los datos
import pandas as pd

trending_data = pd.read_csv('datasets/trending_by_time.csv')
trending_data.head(10)
```

```python
trending_data.sample(10)
```

Voy a realizar un preprocesamiento para formatear adecuadamente la columna de fechas y asegurarme de que los datos estén listos para el análisis en Tableau.

```python
# Convertir la columna 'trending_date' al formato datetime
trending_data['trending_date'] = pd.to_datetime(trending_data['trending_date'], errors='coerce')

# Verificar si hay valores nulos después de la conversión
missing_dates = trending_data['trending_date'].isnull().sum()
print(f'Fechas faltantes después de la conversión: {missing_dates}')

# Verificar algunos registros después de la conversión
trending_data.head()
```

```python
# Guardar el archivo procesado en formato CSV
output_path = 'datasets/trending_by_time_preprocessed.csv'
trending_data.to_csv(output_path, index=False)

print(f'Archivo guardado en: {output_path}')
```

La conversión de la columna trending_date al formato datetime fue exitosa y no se perdieron datos durante el proceso (no hay valores nulos).

El conjunto de datos está listo para ser utilizado en Tableau para crear el dashboard.
